Transformation des données

```python
df_details['DateSortie'] = pd.to_datetime(df_details['yearpublished'], errors='coerce')
```

```python
df_details['DuréeMoy'] = pd.to_timedelta(df_details['playingtime'], unit='m', errors='coerce')
```

```python
df_details['yearpublished'] = pd.to_datetime(df_details['yearpublished'], errors='coerce', format='%Y').dt.date
df_details['playingtime'] = pd.to_timedelta(df_details['playingtime'], unit='m')
```

Remplacement des valeurs manquantes

```python
df_details['description'].fillna('', inplace=True)
df_details['boardgamecategory'].fillna('Inconnu', inplace=True)
df_details['boardgamemechanic'].fillna('Inconnu', inplace=True)
df_details['boardgamefamily'].fillna('Inconnu', inplace=True)
df_details['boardgameexpansion'].fillna('Inconnu', inplace=True)
df_details['boardgameimplementation'].fillna('Inconnu', inplace=True)
df_details['boardgameartist'].fillna('Inconnu', inplace=True)
df_details['boardgamepublisher'].fillna('Inconnu', inplace=True)
df_details['boardgamedesigner'].fillna('Inconnu', inplace=True)
```

```python
df_ratings['thumbnail'].fillna('Indisponible', inplace=True)
```

Fusion des 2 fichiers (dataframes)

```python
df = pd.merge(df_details, df_ratings, on='id', suffixes=('_details', '_ratings'))
```

Création des fichiers pour remplir la base de données

```python
ages = sorted(df["minage"].dropna().unique())
with open("insert_categorie_age.sql", "w") as f:
    for i, age in enumerate(ages, 1):
        f.write(f"INSERT INTO CategorieAge (IDAge, AgeMin) VALUES ({i}, {age});\n")


genres_set = set()
for genres in df["boardgamecategory"].dropna():
    genres_set.update([g.strip() for g in genres.split(",")])
genres_list = sorted(genres_set)

genre_map = {g: i+1 for i, g in enumerate(genres_list)}
with open("insert_categorie_genre.sql", "w") as f:
    for genre, id_genre in genre_map.items():
        genre_sql = genre.replace("'", "''")  
        f.write(f"INSERT INTO CategorieGenre (IDGenre, NomGenre) VALUES ({id_genre}, '{genre_sql}');\n")



mots_cles_set = set()
for mots in df["boardgamemechanic"].dropna():
    mots_cles_set.update([m.strip() for m in mots.split(",")])
mots_cles_list = sorted(mots_cles_set)

motcle_map = {m: i+1 for i, m in enumerate(mots_cles_list)}
with open("insert_motcle.sql", "w") as f:
    for mot, id_mot in motcle_map.items():
        mot_sql = mot.replace("'", "''") 
        f.write(f"INSERT INTO MotCle (IDMotCle, NomMotCle) VALUES ({id_mot}, '{mot_sql}');\n")




with open("insert_jeu.sql", "w", encoding="utf-8") as f:
    for i, row in df.iterrows():
        id_jeu = row['id']
        nom_jeu = row['name'].replace("'", "''") if pd.notnull(row['name']) else f'Jeu_{id_jeu}'
        image = row['thumbnail'] if pd.notnull(row['thumbnail']) else ''
        description = row['description'].replace("'", "''") if pd.notnull(row['description']) else 'Pas de description'
        nb_min = int(row['minplayers']) if pd.notnull(row['minplayers']) else 1
        nb_max = int(row['maxplayers']) if pd.notnull(row['maxplayers']) else 4
        createur = row['boardgamedesigner'].split(',')[0].strip().replace("'", "''") if pd.notnull(row['boardgamedesigner']) else 'Inconnu'
        date_sortie = row['DateSortie'].strftime('%Y-%m-%d') if pd.notnull(row['DateSortie']) else '2000-01-01'
        duree = str(row['DuréeMoy']).split()[2] if pd.notnull(row['DuréeMoy']) else '00:30:00'
        age = int(row['minage']) if pd.notnull(row['minage']) else 6
        id_age = ages.index(age) + 1 if age in ages else 1

        f.write(
            f"INSERT INTO Jeu (IDJeu, NomJeu, Image, Descriptions, NbJoueursMax, NbJoueursMin, Createur, DateSortie, DureeMoy, IDAge) "
            f"VALUES ({id_jeu}, '{nom_jeu}', '{image}', '{description}', {nb_max}, {nb_min}, '{createur}', '{date_sortie}', '{duree}', {id_age});\n"
        )


with open("insert_appartenir.sql", "w") as f:
    for i, row in df.iterrows():
        id_jeu = row['id']
        if pd.notnull(row['boardgamecategory']):
            for genre in [g.strip() for g in row['boardgamecategory'].split(",")]:
                f.write(f"INSERT INTO Appartenir (IDJeu, IDGenre) VALUES ({id_jeu}, {genre_map[genre]});\n")


with open("insert_associer.sql", "w") as f:
    for i, row in df.iterrows():
        id_jeu = row['id']
        if pd.notnull(row['boardgamemechanic']):
            for mot in [m.strip() for m in row['boardgamemechanic'].split(",")]:
                f.write(f"INSERT INTO Associer (IDJeu, IDMotCle) VALUES ({id_jeu}, {motcle_map[mot]});\n")
```
